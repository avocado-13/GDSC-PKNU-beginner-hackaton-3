```python
import pandas as pd
import plotly.graph_objects as go

# CSV 파일 불러오기
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/healthcare_statistics.csv')

# "불안장애 환자수" 데이터 추출 (불안장애 환자 인덱스는 예를 들어 1로 가정)
anxiety_patients = df.loc[4, '2017년':'2021년']

# 데이터 타입을 정수로 변환
anxiety_patients = anxiety_patients.str.replace(',', '').astype(int)

# 데이터프레임으로 변환
anxiety_patients_df = anxiety_patients.reset_index()
anxiety_patients_df.columns = ['연도', '환자수']

# 막대 그래프와 줄 그래프 생성
fig = go.Figure()

# 막대 그래프 추가
fig.add_trace(go.Bar(x=anxiety_patients_df['연도'], y=anxiety_patients_df['환자수'],
                     name='환자 수', marker_color='skyblue'))

# 줄 그래프 추가
fig.add_trace(go.Scatter(x=anxiety_patients_df['연도'], y=anxiety_patients_df['환자수'],
                         name='환자 수', mode='lines+markers', line=dict(color='red')))

# 세로축 범위를 자동 설정 (필요에 따라 수동으로 조정 가능)
fig.update_layout(yaxis=dict(range=[650000, 900000],
                             title='환자 수',
                             tickvals=[650000, 700000, 750000, 800000, 850000, 900000],
                             ticktext=['65만', '70만', '75만', '80만', '85만', '90만']))

# 그래프 제목 및 축 라벨 설정 (한글)
fig.update_layout(title='연도별 불안장애 환자 수 (2017-2021)',
                  xaxis_title='연도')

# 그래프 보여주기
fig.show()
```

```python
import pandas as pd
import plotly.graph_objects as go

# CSV 파일 불러오기
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/healthcare_statistics.csv')

# "우울증 환자수" 데이터 추출
depression_patients = df.loc[0, '2017년':'2021년']

# 데이터 타입을 정수로 변환
depression_patients = depression_patients.str.replace(',', '').astype(int)

# 데이터프레임으로 변환
depression_patients_df = depression_patients.reset_index()
depression_patients_df.columns = ['연도', '환자수']

# 막대 그래프와 줄 그래프 생성
fig = go.Figure()

# 막대 그래프 추가
fig.add_trace(go.Bar(x=depression_patients_df['연도'], y=depression_patients_df['환자수'],
                     name='환자 수', marker_color='skyblue'))

# 줄 그래프 추가
fig.add_trace(go.Scatter(x=depression_patients_df['연도'], y=depression_patients_df['환자수'],
                         name='환자 수', mode='lines+markers', line=dict(color='red')))

# 세로축 범위를 650,000에서 950,000으로 설정
fig.update_layout(yaxis=dict(range=[650000, 950000],
                             title='환자 수',
                             tickvals=[650000, 700000, 750000, 800000, 850000, 900000, 950000],
                             ticktext=['65만', '70만', '75만', '80만', '85만', '90만', '95만']))

# 그래프 제목 및 축 라벨 설정 (한글)
fig.update_layout(title='연도별 우울증 환자 수 (2017-2021)',
                  xaxis_title='연도')

# 그래프 보여주기
fig.show()
```
